b) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).

```python
df['Mileage'] = df['Mileage'].str.replace(r'\s+[a-zA-Z/]+', '', regex=True)
df['Engine'] = df['Engine'].str.replace(r'\s+CC', '', regex=True)
df['Power'] = df['Power'].str.replace(r'\s+bhp', '', regex=True)
df['New_Price'] = df['New_Price'].str.replace(r'\s+Lakh', '', regex=True)

df.head()
```

```python
# replace the missing values in Mileage, Engine, Power and Seats columns
```

```python

import numpy as np

# Convert the columns to numeric, coercing errors into NaN
df['Mileage'] = pd.to_numeric(df['Mileage'], errors='coerce')
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')

# Replace NaN values in numeric columns with their median
df['Mileage'].fillna(df['Mileage'].median(), inplace=True)
df['Engine'].fillna(df['Engine'].median(), inplace=True)
df['Power'].fillna(df['Power'].median(), inplace=True)

# Replace NaN values in the 'Seats' column using the mode
df['Seats'].fillna(df['Seats'].mode()[0], inplace=True)
```

```python

# dropping the 'New_Price' column since there are lot of missing values
```

```python
df.drop('New_Price', axis=1, inplace=True)
```

```python
print(df.isnull().sum())
```

c) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python
fuel_mapping = {'Diesel': 0, 'Petrol': 1, 'Electric': 2}
df['Fuel_Type'] = df['Fuel_Type'].replace(fuel_mapping)

transmission_mapping = {'Manual': 0, 'Automatic': 1}
df['Transmission'] = df['Transmission'].replace(transmission_mapping)

df.head()
```

```python
# getting current year to calculate age of the car
```

```python
from datetime import datetime

current_year = datetime.now().year
```

d) Create one more feature and add this column to the dataset (you can use mutate function in R for this). For example, you can calculate the current age of the car by subtracting “Year” value from the current year.

```python
df['Age_of_Car'] = current_year - df['Year']
```

```python
df['Power_per_Liter'] = (df['Power'].astype(float) / df['Mileage'].astype(float)).round(2)
```

```python
df['Mileage_per_Seat'] = (df['Mileage'].astype(float) / df['Seats'].astype(float)).round(2)
```
